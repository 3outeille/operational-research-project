```python
def adjacency_list_to_adjacency_matrix(adjacency_list):
    n = len(adjacency_list)
    M = [[0 for j in range(n)] for i in range(n)]
    for i in range(n):
        for adj in adjacency_list[i]:
            M[i][adj] = 1
    return M

def pretty_print_adjacency_matrix(adjacency_matrix):
    n = len(adjacency_matrix)
    
    print("  ",end="")
    for i in range(n):
        print("{} ".format(i), end="")
    
    print()
    
    for i in range(n):
        print("{} ".format(i), end="")
        for j in range(n):
            print(adjacency_matrix[i][j], end="|")
        print()

adjacency_matrix = adjacency_list_to_adjacency_matrix(adjacency_list)
pretty_print_adjacency_matrix(adjacency_matrix)
```

```python
def euler_tour(adjacency_matrix, start):
    circuit = []
    return euler_tour_aux(circuit, adjacency_matrix, start)

def euler_tour_aux(circuit, adjacency_matrix, start):
    for i in range(len(adjacency_matrix)):
        if adjacency_matrix[start][i] > 0:
            adjacency_matrix[start][i] -= 1
            adjacency_matrix[i][start] -= 1
            euler_tour_aux(circuit, adjacency_matrix, i)
    circuit.append(start)
    return circuit
        
adjacency_list = dict([(n, list(nbrdict.keys())) for n, nbrdict in MG.adjacency()])
adjacency_matrix = adjacency_list_to_adjacency_matrix(adjacency_list)
circuit = euler_tour(adjacency_matrix, 3)
print(circuit[::-1])
```

```python
def osmnx_to_regular_graph(osmnx_graph, is_oriented=False):
    """
        Convert osmnx graph to regular graph [(src, dest, length)]
        Params:
        Returns:
    """
    if not is_oriented:
        osmnx_graph = ox.utils_graph.get_undirected(osmnx_graph)
        
    osmnx_nodes = list(osmnx_graph.nodes)
    edges = []
    
    for src, dest, _ in list(osmnx_graph.edges):
        length = ox.utils_graph.get_route_edge_attributes(osmnx_graph,
            (src, dest), 'length')[0]
        edges.append((osmnx_nodes.index(src), osmnx_nodes.index(dest), length))
        
    return len(osmnx_nodes), edges

nodes, edges = osmnx_to_regular_graph(MG)

for src, dest, length in edges:
    print(src, dest, length)
```
