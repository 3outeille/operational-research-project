```python
%matplotlib inline
import matplotlib.pyplot as plt
import osmnx as ox
import networkx as nx
from scipy import sparse
from itertools import combinations
import random
```

```python
filename = "maps/directed-weighted-graph-10.npz"
G_sparse = sparse.load_npz(filename)
G = sparse.csgraph.csgraph_to_dense(G_sparse)

# Plot graph
fig, ax = plt.subplots(figsize=(10,10))
graph = nx.convert_matrix.from_numpy_array(G)
nx.draw_networkx(graph, ax=ax, with_labels=True)

# Weights
weights = nx.get_edge_attributes(graph, "weight")
weights
```
